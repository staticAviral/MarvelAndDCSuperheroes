# Loading required libraries

```python
import pandas as pd
from matplotlib import pyplot as plt
import seaborn as sns
```

```python
marvel=pd.read_csv('charcters_stats.csv')
```

```python
marvel.head()
```

```python
marvel.shape
```

```python
marvel['Alignment'].value_counts()
```

```python
good=marvel[marvel['Alignment']=='good']
```

```python
good.head()
```

# Top 5 Fastest Superheroes

```python
good.sort_values(by=['Speed'],ascending=False)[0:5]
```

# Top 5 Powerful Superheroes

```python
good.sort_values(by=['Power'],ascending=False).head()
```

```python
good_max_power_full=good[good['Power']==100]
```

# Superheroes with power=100

```python
good_max_power_full
```

```python
good_max_power=good.sort_values(by=['Total'],ascending=False)
```

# Top 5 Superheroes w.r.t. total

```python
good_max_power.head()
```

# Bar plot for top 5 superheroes w.r.t total

```python
plt.figure(figsize=(7,6))
plt.bar(list(good_max_power['Name'])[0:5],list(good_max_power['Total'])[0:5],color='blue')
plt.show()
```
